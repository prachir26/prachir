### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your answer here
#First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list
movie_item_data = movie_item.merge(movie_data, left_on='movie id', right_on = 'movie id')
movie_item_data_grp =  movie_item_data.groupby(['movie title']).count()
movie_item_data_grp = movie_item_data_grp.query('rating > 100')
movie_list = movie_item_data_grp.index

#Find the average rating of all the movies and sort them in the descending order. 
#You will have to use the .merge() function to reach to a data set through which you can get the 
#names and the average rating.
movie_item_data_grp1 = movie_item_data.groupby(['movie title']).mean()
movie_item_data_grp1 = movie_item_data_grp1.sort_values(by='rating', ascending=False)

#Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

movie_item_data_grp2 = movie_item_data_grp1[movie_item_data_grp1.index.isin(movie_list)]

movie_item_data_grp2.head(25)
```
